```python
import json
import pandas as pd

with open("haul.json", "r") as file:
    data = json.load(file)

tags = []
requests = []
responses = []

for haul in data["haul"]:
    tag = haul["tag"]
    for request in haul["request"]:
        requests.append(request)
        tags.append(tag)
        responses.append(haul["response"][0])

dataset = pd.DataFrame({
    "tags": tags,
    "request": requests,
    "responses": responses
})

dataset.to_csv("dataset.csv", index=False)
```

```python
# importing spacy for loading en_core_web_md 
import spacy

# importing nltk for tokenizing, lemmatizing
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# SVM for SVC(kernel="linear") model=2
from sklearn.svm import SVC

# importing warning for filterwarning
import warnings
warnings.filterwarnings("ignore")

# labelencoder for deeplearning
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer

# tensorflow model 
import tensorflow as tf


# global initialization
nlp = spacy.load("en_core_web_md")
lemmatizer = WordNetLemmatizer()

dataset = pd.read_csv("dataset.csv")

dataset.head()
```

```python
def lammetizing(phrase):
    words = word_tokenize(phrase)
    lemmatizing_words = []
    for word in words:
        lemmatizing_words.append(lemmatizer.lemmatize(word))
    return " ".join(lemmatizing_words).lower()

x_train = []
y_train = dataset["tags"]

for phrase in dataset["request"]:
    x_train.append(lammetizing(phrase))

pd.concat([pd.Series(x_train[:10]), pd.Series(y_train[:10])], axis=1)
```

```python
# x_train as vector format
docs = [nlp(text) for text in x_train]
x_train_vector = [doc.vector for doc in docs]

y_train_vector = []

tags = ["greetings", "aboutyou", "aboutme", "thanks", "goodbye", "arulkumaran", "ambition"]

for tag in y_train:
    if tag == "greetings":
        y_train_vector.append(0)
    elif tag == "aboutyou":
        y_train_vector.append(1)
    elif tag == "aboutme":
        y_train_vector.append(2)
    elif tag == "thanks":
        y_train_vector.append(3)
    elif tag == "goodbye":
        y_train_vector.append(4)
    elif tag == "arulkumaran":
        y_train_vector.append(5)
    elif tag == "ambition":
        y_train_vector.append(6)
    else:
        continue
```

```python
# SVM model
model = SVC(kernel="linear")

# fit the model
model.fit(x_train_vector, y_train_vector)
```

```python
x_train = dataset["request"]
y_train = dataset["tags"]

y = []

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y_train)

vectorizer = CountVectorizer()
x = vectorizer.fit_transform(x_train).toarray()

x[2], y[2]
```

```python
# Deeplearning model
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(90, activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(50, activation="relu"),
    tf.keras.layers.Dense(len(dataset), activation="softmax")
])

# compile the model
model1.compile(loss=tf.keras.losses.sparse_categorical_crossentropy,
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              metrics=["accuracy"])

# fir the model
model1.fit(x, y, epochs=500)
```

```python
# model prediction
prediction = model1.predict(vectorizer.transform(["say your ambition"]).toarray())
intent = label_encoder.inverse_transform([prediction.argmax()])

intent
```

```python
model1.save("haul.h5")
```
